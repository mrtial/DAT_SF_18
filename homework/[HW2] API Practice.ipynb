```python
# I defined a function get_set() to convert string 'TIME' into 'Sec'(second), 
# so I can compare different time.
def get_sec(s):
    l = s.split(':')
    return int(l[0]) * 3600 + int(l[1]) * 60 + int(l[2])
```

```python
# generate another column 'Sec'
# This is stupid, I konw. But I don't know how to do it faster.

df2 = df
df2['Sec']=df['TIME']

S=[]
for s in df2['Sec']:
    s=get_sec(s)
    S.append(s)

df2['NO']=df['NO']

N=[]
for n in df2['NO']:
    N.append(n)

D=dict(zip(N, S))
print D
```
