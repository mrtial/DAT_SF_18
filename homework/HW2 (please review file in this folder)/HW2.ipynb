### Preparation: Generate 'Sec', 'Gender' & 'Group'
- Time is string, cannot be used for plotting and comparing. 
- Division conatins 'gender' and 'age groupe' information.

```python
# Create new column with value of 'DIVISION' and replace it with part of the string value. 
df['Gender']=df['DIVISION']
df['Group']=df['DIVISION']

for i in range(len(df.Gender)): 
    df.Gender[i]=str(df.Gender[i])[0]
    df.Group[i]=str(df.Group[i])[1:5]
```

##### The result looks good!

```python
# defined a function called 'get_sec()' to turn 'TIME' string into second
def get_sec(s):
    l = s.split(':')
    return int(l[0]) * 3600 + int(l[1]) * 60 + int(l[2])
```

```python
# convert the value of 'TIME' into 'Sec'.
df['Sec']=df['TIME']
for i in range(len(df.Sec)):
    df.Sec[i]=get_sec(df.TIME[i])
```

```python
# Group my data with 'Group':
grouped = df.groupby('Group')
```

```python
# Generate runner performance by age division.

g1=list(grouped.get_group('0119').Sec)
g2=list(grouped.get_group('2029').Sec)
g3=list(grouped.get_group('3039').Sec)
g4=list(grouped.get_group('4049').Sec)
g5=list(grouped.get_group('5059').Sec)
g6=list(grouped.get_group('6069').Sec)
g7=list(grouped.get_group('70UP').Sec)

d = {"0-19":g1, "20-29":g2, "30-39":g3, "40-49":g4, "50-59":g5, "60-69":g6,"70up":g7}
```

```python
dfg1=pd.DataFrame(g1) 
dfg2=pd.DataFrame(g2)
dfg3=pd.DataFrame(g3)
dfg4=pd.DataFrame(g4)
dfg5=pd.DataFrame(g5)
dfg6=pd.DataFrame(g6)
dfg7=pd.DataFrame(g7)

group_data=[dfg1,dfg2,dfg3,dfg4,dfg5,dfg6,dfg7]
```

```python
# Boxplot for different group:

fig, ax1 = plt.subplots(figsize=(10, 6))
fig.canvas.set_window_title('Golden Gate Half 2015')
plt.subplots_adjust(left=0.075, right=0.95, top=0.9, bottom=0.25)

bp = plt.boxplot(group_data, notch=0, sym='+', vert=1, whis=1.5)

plt.setp(bp['boxes'], color='black')
plt.setp(bp['whiskers'], color='black')
plt.setp(bp['fliers'], color='red', marker='+')
plt.xticks(range(8), [' ','01-19','20-29','30-39','40-49','50-59','60-69','Over 70'])

ax1.yaxis.grid(True, linestyle='-', which='major', color='lightgrey',
               alpha=0.5)

# Hide these grid behind plot objects
ax1.set_axisbelow(True)
ax1.set_title('Runner Performance Across Different Age Group')
ax1.set_xlabel('DIVISION')
ax1.set_ylabel('TIME/Sec')

plt.show()
```
